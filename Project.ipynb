```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Diwali Sales Data.csv',encoding='unicode_escape')
# to avoid UnicodeDecode Error encoding is used
```

```python
df.shape
# to chk whether file has loaded or not
```

```python
df.head()
# to see top 5 data
```

```python
df.info()
# to get information about table
```

```python
# DATA CLEANING
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
# drop/delete unrelated/ blank columns as can be seen above(from info)
# axis=1 means whole row
# inplace=True make changes permanent in table
```

```python
df.info()
# to chk whether status and unnamed is deleted or not
```

```python
pd.isnull(df)
# to see whether value is null or not 
# false means not NULL and true means NULL
```

```python
pd.isnull(df).sum()
#from above data we can only see but if we want sum of how many null vlues we add sum
```

```python
# above data tells that there is 12 null values in amount column
```

```python
df.dropna(inplace=True)
# to delete null values from table
```

```python
pd.isnull(df).sum()
# to chk whther Null values are deleted or not
```

```python
# EXPLORATORY DATA ANALYSIS
```

```python
# GENDER
ax=sns.countplot(x='Gender',data=df)
```

```python
# this graph tells classification based o number of female and male
ax=sns.countplot(x='Gender',data=df)
# to show values above bar for loop is used
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# to know who spend more in female or male
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount')
```

```python
sales=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales)
```

```python
From above graph we can see that most of the buyers are Female and even purchasing power of female is more
```

```python
# AGE
ax=sns.countplot(x='Age Group',data=df,hue='Gender')
# hue is used to differentiate btw male and female
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales)
```

```python
From the above graph it is clear that most of the buyers are of age group btw 26-35 yrs female
```

```python
# STATE
# total number of orders from state
sales=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)
# since names of state is not clerly visible set the size of plot
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='State',y='Orders',data=sales)
```

```python
# total amout from each state spend on buying
sales=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
# since names of state is not clerly visible set the size of plot
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='State',y='Amount',data=sales)
```

```python
From above graph we can see that most of the orders are from Uttar Pradesh, Maharashtra and Karnataka 
but by total sales or amount is from Uttar Pradesh , Karnataka and Maharashtra 
```

```python
# MARITAL STATUS

ax=sns.countplot(x='Marital_Status',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Marital_Status',y='Amount',data=sales,hue='Gender')
```

```python
From the above graph we can say that most of the buyers are married (women) and they have high purchasing power
```

```python
# OCCUPATION

ax=sns.countplot(x='Occupation',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Occupation',y='Amount',data=sales)
```

```python
From the above graph it can be seen that most of the buyers are from IT Sector
```

```python
# PRODUCT CATEGORY

ax=sns.countplot(x='Product_Category',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Product_Category',y='Amount',data=sales)
```

```python
From the above graph it can be seen that most of the sold products are of Food 
```

```python
# CONCLUSION
Married omen age group from 26-35 years from Uttar Pradesh working in IT Sector are more likely to buy food products.
```
